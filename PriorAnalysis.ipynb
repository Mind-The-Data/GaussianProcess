```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import pymc3 as pm
```

#### Gamma dist for eta and rho

```python
# eta param
# I like 5, 10. 
plt.style.use('seaborn-darkgrid')
x = np.linspace(0, 3, 200)
alphas = [5, 4, 3, 6]
betas = [10, 8, 6, 12]
for a, b in zip(alphas, betas):
    pdf = st.gamma.pdf(x, a, scale=1.0/b)
    plt.plot(x, pdf, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

```python
# lengthscale param ls
# 100-1000
# I like 2, .007
plt.style.use('seaborn-darkgrid')
x = np.linspace(0, 1000, 200)
alphas = [3, 2, 5, 6]
betas = [.01, .007, .01, .015]
for a, b in zip(alphas, betas):
    pdf = st.gamma.pdf(x, a, scale=1/b, loc=0)
    plt.plot(x, pdf, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

```python
# lengthscale param lsq
# 100**2-1000**2
# a=3.5, b=1.2e-5
plt.style.use('seaborn-darkgrid')
x = np.linspace(0, 800**2, 200)
alphas = [4, 3, 3.5, 10]
betas = [.00007, .00001, .000012, .00005]
for a, b in zip(alphas, betas):
    pdf = st.gamma.pdf(x, a, scale=1/b, loc=0)
    plt.plot(x, pdf, label=r'$\alpha$ = {}, $\beta$ = {}'.format(a, b))
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(loc=1)
plt.show()
```

```python
#ls = np.linspace(0,1000,100)
#plt.plot(ls, 1*np.exp(1e4/ls));

for esq,lsq in zip([1,1,1,.3],[1e4,4e4,1e5,5e5]):
    ls = round(np.sqrt(lsq))
    #es = round(np.sqrt(esq),1)
    x = np.linspace(0,1e3,50)
    plt.plot(x,esq*np.exp(-x**2/(lsq)), label=r'$3ls$ = {}, $\sigma^2$ = {}'.format((3*ls),(esq)))

plt.xlabel('distance (miles)')
plt.legend();
#plt.title(r'  $l$ = {}'.format(ls) + r',   $3l$ = {}'.format(3*ls));
```
